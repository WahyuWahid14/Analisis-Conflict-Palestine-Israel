```python
data = pd.read_csv("C:\\Users\\Wahyu Wahid\\Documents\\Latihan Analisis Bisnis\\conflict_palestin_israel.csv")
data.head()
# data.info()
```

```python
data['date_of_death'] = pd.to_datetime(data['date_of_death'])
data['date_of_event'] = pd.to_datetime(data['date_of_event'])
```

#### NO.1
###### Kita bisa lihat dari line chart di bawah berdasarkan table di bawahnya bahwasannya jumlah kematian dari tahun 2000 - 2023 itu terjadi inflasi dan deflasi terhadap jumlah kematian yang di alami dalam peristiwa "Conflict Palestine Dan Israel". Kemudian kita juga dapat melihat inflasi tertinggi  dalam konflik ini itu pada tahun 2014. Kalau kita lihat pada tahun tersebut itu terjadi peristiwa konflik di Jalur Gaza antara Palestina dan Israel, yang dikenal sebagai Operasi Protective Edge. Selama konflik tersebut, korban jiwa di pihak Palestina mencapai ribuan, dengan jumlah yang disebutkan sekitar 2.326 orang.Dari data yang diberikan, terlihat adanya fluktuasi jumlah kematian akibat konflik antara Palestina dan Israel dari tahun 2000 hingga 2023. Pada beberapa tahun, seperti 2002 dan 2014, terjadi lonjakan signifikan dalam jumlah kematian, mungkin terkait dengan eskalasi konflik intens. Namun, pada tahun-tahun lain, terlihat penurunan atau angka yang relatif rendah.

```python
result_df = data.groupby(data['date_of_death'].dt.year)['date_of_event'].count().reset_index()

result_df.columns = ['tahun', 'jumlah_kematian']
plt.figure(figsize=(10, 6))
plt.plot(result_df['tahun'], result_df['jumlah_kematian'], marker='o', linestyle='-', color='b')
plt.title('Jumlah Kematian Berdasarkan Tahun')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Kematian')
plt.grid(True)
plt.show()
result_df.head(30)
```

#### NO.2


```python
# age
data['kelompok_usia'] = pd.cut(data['age'],
                             bins=[0, 12, 18, 50, float('inf')],
                             labels=['Anak-anak', 'Remaja', 'Dewasa Pertengahan', 'Lansia'])
data_usia = data.groupby('kelompok_usia').agg({'name': 'count'}).reset_index()
labels_usia = data_usia['kelompok_usia'] 
sizes_usia = data_usia['name']

# gender
gender_df = data.groupby('gender').size().reset_index(name='jumlah_kematian')
gender_df = gender_df.sort_values(by='jumlah_kematian', ascending=False)
labels_gender = gender_df['gender']
sizes_gender = gender_df['jumlah_kematian']

# citizenship
citizenship_df = data.groupby('citizenship').size().reset_index(name='jumlah_kematian_individu')
citizenship_df = citizenship_df.sort_values(by='jumlah_kematian_individu', ascending=False)

# Plotting
fig, axs = plt.subplots(1, 3, figsize=(18, 6))

# Plot untuk Usia
axs[0].pie(sizes_usia, labels=labels_usia, autopct='%1.2f%%')
axs[0].set_title('Distribusi Usia')
axs[0].legend(labels_usia, loc='upper right')

# Plot untuk Jenis Kelamin
axs[1].pie(sizes_gender, labels=labels_gender, autopct='%1.2f%%')
axs[1].set_title('Distribusi Jenis Kelamin')
axs[1].legend(labels_gender, loc='upper right')

# Plot untuk Kewarganegaraan
bar_container = axs[2].bar(citizenship_df['citizenship'], citizenship_df['jumlah_kematian_individu'])
axs[2].set(ylabel='Jumlah Kematian', title='Jumlah Kematia Berdasarkan Kewarganegaraan', xlabel='Kota')
axs[2].bar_label(bar_container, fmt='{:,.0f}')
# axs[2].legend()

plt.tight_layout()
plt.show()
```

###### Kalau di lihat berdasarkan jenis cidera kemudian penyebab kematiannya(yang membunuh siapa), kita bisa melihat bahwasannya yang tingkat pembunuhan yang di lakukan israel terhadap palestina itu sangat tinggi, dan pembunuhan yang di lakukan oleh warga sipil palestin jauh sangant rendah. Dari sini kita dapat melihat partisipasi Israel di dominasi dengan cara membunuh yaitu dengan "Tembakan" yang di lakukan oleh Pasukan keamanan israel, Sedangkan Warga sipil palestina membunuh melakukan dengan cara dengan cara "Ledakan".

```python
result =data.groupby(['type_of_injury', 'killed_by']).size().reset_index(name='jumlah_kasus_kematian')

result = result.sort_values(by='jumlah_kasus_kematian', ascending=False)

result.columns = ['type_of_injury', 'killed_by', 'jumlah_kasus_kematian']
# result.info()
# Bar Chart
plt.figure(figsize=(10, 10))
sns.barplot(x='jumlah_kasus_kematian', y='type_of_injury', hue='killed_by', data=result)
plt.title('Jumlah Kasus Kematian Berdasarkan Tipe Cedera, Penyebab Kematian')
plt.show()
# result.head(30)
```

```python
# age
data['kelompok_usia'] = pd.cut(data['age'],
                             bins=[0, 12, 18, 50, float('inf')],
                             labels=['Anak-anak', 'Remaja', 'Dewasa Pertengahan', 'Lansia'])
data_usia = data.groupby('kelompok_usia').agg({'name': 'count'}).reset_index()
labels_usia = data_usia['kelompok_usia'] 
sizes_usia = data_usia['name']

# gender
gender_df = data.groupby('gender').size().reset_index(name='jumlah_kematian')
gender_df = gender_df.sort_values(by='jumlah_kematian', ascending=False)
labels_gender = gender_df['gender']
sizes_gender = gender_df['jumlah_kematian']

# place_of_residence_district
result = data['place_of_residence_district'].value_counts().reset_index()
result.columns = ['daerah_tempat_tinggal', 'jumlah_kematian']
label_jumlah = result['jumlah_kematian']
label_daerah = result['daerah_tempat_tinggal']

fig, axs = plt.subplots(1, 3, figsize=(18, 6))
print(fig)
# Visualisasi distribusi usia
axs[0].bar(labels_usia, sizes_usia, color = 'skyblue')
axs[0].set(xlabel = 'Kategori Usia', title = 'Distribusi Usia', ylabel = 'Total Jumlah')

# Visualisasi persebaran jenis kelamin
axs[1].pie(sizes_gender, labels=labels_gender, autopct='%1.2f%%', colors=['lightcoral', 'lightblue'] )
axs[1].set_title('Distribusi Jenis Kelamin')
axs[1].legend(labels_gender, loc='upper right')

# # Visualisasi lokasi daerah tempat tinggal
axs[2].barh(label_daerah, label_jumlah, color = 'lightgreen')
axs[2].set_title('Lokasi Daerah Tempat Tinggal')

plt.tight_layout() 
plt.show()
```
